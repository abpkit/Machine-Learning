```python
# importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Importing data
data = pd.read_csv("C:/Users/Abhi/Desktop/Advertisement CSV fie.csv")

sns.lineplot(x="TV", y="sales", data = data)            # Column names has to be mentioned in " ".
#sns.show()        seaborn hasno attribute "show"-----> This gives error.
plt.show()
```

```python
sns.barplot(x="TV", y="sales", data=data)
plt.show()
```

```python
sns.scatterplot(x="TV",y="sales", data=data)
plt.show()
```

```python
data['sales'].plot(kind='hist')
plt.xlabel('Advertise on TV')
plt.ylabel('Sales')
plt.show()
```

```python
# to diffeerentiate the sales e.r.t colour we can assign hue=sales
sns.scatterplot(x="TV",y="sales", hue="sales",data=data)
```

```python
sns.jointplot(x="TV",y="sales",data=data)
plt.show()
```

```python
sns.jointplot(x="TV",y="sales",kind="reg",data=data)
plt.show()
```

```python
plt.boxplot(x="sales", data=data)
plt.show()
```

```python
plt.boxplot(x="TV", data=data)
plt.show()
```

```python
sns.boxplot(x="TV", y="sales",data=data)
plt.show()
```

```python
sns.regplot(x="TV", y="sales", data=data)
plt.show()
```

```python
sns.heatmap(data.isnull())
plt.show()
```

```python
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix,annot=True)
plt.show()
```
