```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model
```

```python
diabetes = datasets.load_diabetes()
print(diabetes.keys())
```

```python
diabetes.data
```

```python
print(diabetes.DESCR)
```

```python
diabetes_x = diabetes.data[:, np.newaxis,2]        # 2 is index number
print(diabetes_x)
```

```python
# now we will take 30 sample from the dataset
diabetes_x_train = diabetes_x[:-30]                  # last 30 samples
diabetes_x_test = diabetes_x[-30:]                   # first 30
diabetes_y_train = diabetes.target[:-30]
diabetes_y_test = diabetes.target[-30:]
```

```python
model = linear_model.LinearRegression()
model.fit(diabetes_x_train, diabetes_y_train)
diabetes_y_predict = model.predict(diabetes_x_test)
```

```python
from sklearn.metrics import mean_squared_error
print("Mean sqqured error is :", mean_squared_error(diabetes_y_test, diabetes_y_predict))
```

```python
print("Weights", model.coef_)
print("Intercept", model.intercept_)
```

```python
plt.scatter(diabetes_x_test, diabetes_y_test)
plt.show()
```

```python
plt.plot(diabetes_x_test, diabetes_y_predict)
plt.show()
```

```python
plt.scatter(diabetes_x_test, diabetes_y_test)
plt.plot(diabetes_x_test, diabetes_y_predict)
plt.show()
```
